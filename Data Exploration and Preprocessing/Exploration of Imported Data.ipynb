## Now how does the price of the house change w.r.t. the area of the house

## What if we wanted to find the age distribution of a house. Like number of houses between 3 to 15 years of age, 15 to 27 and so on... 

```python
# Then we'll use a histogram 

# To create a histogram, we divide the entire data in bins. Then these bins divide the entire range of the dataset into several parts. (equal parts)

plt.hist(Raw_Housing_Data['Age of House (in Years)'], bins=10)
plt.xlabel('Age of House (in Years)')
plt.ylabel('Number of Houses')
plt.title('Number of House V/S Age of House (in Years)')
```

# Now we are done with the Graphical Representation of Data, We must now head towards some Advanced numpy and pandas funcitons

```python
# How Do I find how Condition of the House Affects the sale price of the house. 
# Now this is easy right. NO! We are now connecting a Continious Variable with a Categorical Variable. Now this is capable of lynching some boobs

Raw_Housing_Data['condition_sale'] = 0

for i in Raw_Housing_Data['Condition of the House'].unique():
    Raw_Housing_Data['condition_sale'][Raw_Housing_Data['Condition of the House'] == str(i)] = Raw_Housing_Data['Sale Price'][Raw_Housing_Data['Condition of the House'] == str(i)].mean()
    
    
    
    
```

```python
# Plotting the mean sales based upon the Condition of the House

plt.figure(dpi = 100)
plt.bar(x=Raw_Housing_Data['Condition of the House'], y=Raw_Housing_Data['condition_sales'])
plt.xlabel('Condition of the House')    
plt.ylabel('Average Sale Price of the House')
plt.title('Average Sale Price V/S Condition of the House')
```
